```python
시작 = 1
끝 = 10
while 시작 < 끝:
    if 시작%2 == 0:
       print(f"출력 : {시작}단")
       
       
       곱 = 1
       while 곱 < 끝:
           if 곱%2  != 0:
               print(f"{시작} * {곱} = {시작 * 곱}")
           곱 = 곱 + 1
            
       
    시작 = 시작 + 1
```

곱이 1씩 증가하는 기존 로직은 그대로 두면서 특정 코드를 추가하는 방식으로 곱셈 결과가 홀수일 때만 출력되도록 만들 수 있습니다.

앞서 시작 변수가 '짝수'일 때만 단 전체가 출력되도록 만들었던 방법을 떠올려 보세요. 그것과 아주 유사한 원리를 안쪽 while문에도 적용할 수 있습니다.

```python
선택한음료 = input("콜라,사이다,홍차 중 하나를 고르세요")
금액 = 0
while 선택한음료:
    if 선택한음료 == "콜라":
      금액 = 500
    elif 선택한음료 == "사이다":
      금액 = 550
    elif 선택한음료 == "홍차":
      금액 = 1500

    수량 = int(input("수량을 입력하세요."))
    print(f'{선택한음료} {수량}개의 금액은 {금액 * 수량}입니다.')
    지불한금액 = int(input("지불할 금액을 알려주세요."))
    print(f'{지불한금액}원을 받고 {지불한금액-(금액*수량)}원 거스름돈을 받으세요')
    
    
    선택한음료 = input("anything else? (콜라,사이다,홍차 중 택 1. 종료 Enter) ")
    

print("감사합니다.")
```
